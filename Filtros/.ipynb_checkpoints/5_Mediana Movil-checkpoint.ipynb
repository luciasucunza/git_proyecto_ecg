Each signal was processed with a
median filter of 200-ms width to remove QRS complexes and
P-waves. The resulting signal was then processed with a median
filter of 600 ms width to remove T-waves. The signal resulting
from the second filter operation contained the baseline
of the ECG signal, which was then subtracted from the original

Además dependiendo de la base de datos la frecuencia de 30, 50 o 60 cambia segun la base de datos o la señal que se usa, por eso es necesario conocer la fft de la señal que estoy tratando, en este caso la linea es de 50Hz

```python
import warnings
warnings.filterwarnings('ignore')

import numpy as np
import matplotlib.pyplot as plt
import scipy.io as sio
import scipy.signal as sig
import time


#------APERTURA DE LA SEÑAL-------
mat_struct = sio.loadmat('/home/luciasucunza/git_proyecto_ecg/Filtros/TP4_ecg.mat')

ecg_one_lead  = mat_struct['ecg_lead']
ecg_one_lead  = ecg_one_lead.flatten(1)
cant_muestras = len(ecg_one_lead)

fs       = 1000 
nyq_frec = fs / 2
t        = np.arange(cant_muestras) / fs
```

```python
#------OBTENCION DE LA BASELINE-------
to_todo = time.time()

baseline = sig.medfilt (ecg_one_lead, kernel_size = int (np.around(fs*0.5*0.2)    *2 +1  ))
baseline = sig.medfilt (baseline, kernel_size = int (np.around(fs*0.5*0.6)    *2 +1  )) 
baseline = sig.medfilt (baseline, kernel_size = int (np.around(fs*0.5*(1/50)) *2 +1  ))

#------OBTENCION DE LA SEÑAL FILTRADA-------
ecg_F   = ecg_one_lead - baseline

tf_todo = time.time()

#------PLOTEO DE LA SEÑAL FILTRADA-------
plt.figure(2)
plt.title('Señales')
plt.plot(t, ecg_one_lead,       label='ECG sin flitrar')
plt.plot(t, ecg_F,              label='ECG flitrarado' )
plt.plot(t, baseline,           label='Baseline'       )
plt.grid()
plt.legend()
plt.show()
```

Podemos ver que el filtrado es muy bueno, pero tiene un costo computacional muy alto. Este se puede reducir actuando solo en algunas regiones

```python
#------REGIÓN DE TRABAJO-------
to_zoom = time.time()

c_muestras_zoom = 60000
zoom_region     = np.arange(0, c_muestras_zoom , 1)
ecg_zoom        = ecg_one_lead[zoom_region]
t_zoom          = t[zoom_region]

#------OBTENCION DE LA BASELINE-------
baseline_zoom = sig.medfilt (ecg_zoom, kernel_size = int (np.around(fs*0.5*0.2)    *2 +1  ))
baseline_zoom = sig.medfilt (baseline_zoom, kernel_size = int (np.around(fs*0.5*0.6)    *2 +1  )) 
baseline_zoom = sig.medfilt (baseline_zoom, kernel_size = int (np.around(fs*0.5*(1/50)) *2 +1  ))

#------OBTENCION DE LA SEÑAL FILTRADA-------
ecg_F_zoom = ecg_zoom - baseline_zoom

tf_zoom = time.time()

#------PLOTEO DE LA SEÑAL FILTRADA-------
plt.figure(2)
plt.title('Señales con Zoom (0s a ' + str(c_muestras_zoom/fs) + 's)' )
plt.plot(t_zoom, ecg_zoom,       label='ECG sin flitrar')
plt.plot(t_zoom, ecg_F_zoom,     label='ECG flitrarado' )
plt.plot(t_zoom, baseline_zoom,  label='Baseline'       )
plt.grid()
plt.legend()
plt.show()
```

Así podemos comparar como se reduce el tiempo de procesamiento si se reduce la cantidad de muestras a procesar

```python
dt_todo = round(tf_todo - to_todo,2) 
dt_zoom = round(tf_zoom - to_zoom,2)

print('Tiempo de procesamiento de toda la señal:')
print('           ' + str(dt_todo))
print('Tiempo de procesamiento de la region (0s a ' + str(c_muestras_zoom/fs) + 's)')
print('           ' + str(dt_zoom))   
```

Reducir la zona donde se va a aplicar no tiene un sentido practico, pero es necesario reducir la cantidad de muestras que se deben estimar para poder obtener un algoritmo que sea de mayor utilidad. Por eso en el siguiente notbook se utilizará el metodo Multirate para reducir el costo computancional.
